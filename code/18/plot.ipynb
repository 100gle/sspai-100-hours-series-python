```python
import pandas as pd
```

```python
titanic = pd.read_csv("../../data/titanic.csv")
```

```python
titanic.head()
```

```python
titanic["Sex"].value_counts().plot.bar()
```

```python
ax = titanic["Sex"].value_counts().plot.bar()
ax.set_title("The Bar Chart of Sex")
ax.set_xlabel("Sex")
ax.set_ylabel("Counter")
ax.set_xticklabels(["male", "female"], rotation=0)
```

```python
titanic["Sex"].value_counts().plot.pie()
```

```python
titanic["Sex"].value_counts().plot.bar(
    figsize=(8, 6),
    title="The Bar Chart of Sex",
    rot=45,
    xlabel="Sex",
    ylabel="Number",
    ylim=(0, 550),
    legend=True,
    fontsize=8.5,
)
```

```python
from matplotlib import image
from matplotlib.offsetbox import OffsetImage


logo = image.imread("../../data/python-logo-master-v3-TM.png")

ax = titanic["Sex"].value_counts().plot.pie(
    figsize=(6, 8),
    ylabel="",
    title="The Pie Chart of Sex",
    explode=(0, 0.05),
    autopct=lambda pct: f"{round(pct, 2)}%",
    pctdistance=0.4,
    shadow=True,
    colors=("steelblue", "orange"),
)

for n, tx in enumerate(ax.texts):
    if (n+1) % 2==0:
        tx.set_color("white")

logo_image = OffsetImage(logo, zoom=.5)
logo_image.set_offset((35, 25))
ax.add_artist(logo_image)
```
